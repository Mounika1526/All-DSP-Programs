```python
import matplotlib.pyplot as plt
country=["India","Usa","Germany","UK","France"]
GDP_Per_Capita=[48000,38000,50000,45000,44000]
plt.bar(country,GDP_Per_Capita)
plt.show()
```

```python
country=["India","Usa","Germany","UK","France"]
GDP_Per_Capita=[48000,38000,50000,45000,44000]
plt.bar(country,GDP_Per_Capita)
plt.xlabel('country')
plt.ylabel('GDP Per Capita')
plt.show()
```

```python
 country=["India","Usa","Germany","UK","France"]
GDP_Per_Capita=[48000,38000,50000,45000,44000]
plt.bar(country,GDP_Per_Capita)
plt.title("Country vs GDP Per Capita")
plt.xlabel('country')
plt.ylabel('GDP Per Capita')
plt.show()
```

```python
country=["India","Usa","Germany","UK","France"]
GDP_Per_Capita=[48000,38000,50000,45000,44000]
plt.bar(country,GDP_Per_Capita,color="hotpink")
plt.title("Country vs GDP Per Capita")
plt.xlabel('country')
plt.ylabel('GDP Per Capita')
plt.show()
```

```python
country=["India","Usa","Germany","UK","France"]
GDP_Per_Capita=[48000,38000,50000,45000,44000]
plt.barh(country,GDP_Per_Capita,color="hotpink")
plt.title("Country vs GDP Per Capita")
plt.xlabel('country')
plt.ylabel('GDP Per Capita')
plt.show()
```
