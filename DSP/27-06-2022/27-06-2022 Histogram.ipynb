## Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
numbers=[0.1,0.5,1,1.5,2,4,5.5,6,8.8,9]
plt.hist(numbers)
plt.xlabel="Numbers"
plt.ylabel="Frequency"
plt.show()
```

```python
numbers=[0.1,0.5,1,1.5,2,4,5.5,6,8.8,9]
plt.hist(numbers,bins=3)
plt.xlabel="Numbers"
plt.ylabel="Frequency"
plt.show()
```

```python
numbers=[0.1,0.5,1,1.5,2,4,5.5,6,8.8,9]
plt.hist(numbers,bins=2)
plt.xlabel="Numbers"
plt.ylabel="Frequency"
plt.show()
```

```python
numbers=[0.1,0.5,1,1.5,2,4,5.5,6,8.8,9]
plt.hist(numbers,bins=5)
plt.xlabel="Numbers"
plt.ylabel="Frequency"
plt.show()
```

```python
numbers=[0.1,0.5,1,1.5,2,4,5.5,6,8.8,9]
plt.hist(numbers,bins=3,color="r")
plt.xlabel="Numbers"
plt.ylabel="Frequency"
plt.show()
```

```python
numbers=[0.1,0.5,1,1.5,2,4,5.5,6,8.8,9]
plt.hist(numbers,bins=3,color="hotpink")
plt.xlabel="Numbers"
plt.ylabel="Frequency"
plt.show()
```
