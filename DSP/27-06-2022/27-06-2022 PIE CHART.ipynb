## PIE CHARTS

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
plt.pie(y)
plt.show()
```

```python
y=np.array([35,50,15])
co=["r","b","hotpink"]
plt.pie(y,colors=co,labels=["DSP","CD","WT"])
plt.show()
```

```python
y=np.array([35,50,15])
co=["r","b","hotpink"]
plt.pie(y,colors=co,labels=["DSP","CD","WT"])
plt.legend()
plt.show()
```

```python
y=np.array([35,50,20])
co=["r","b","hotpink"]
plt.pie(y,colors=co,labels=["DSP","CD","WT"],startangle=180)
plt.legend(title="subjects",loc="lower right")
plt.show()
```
