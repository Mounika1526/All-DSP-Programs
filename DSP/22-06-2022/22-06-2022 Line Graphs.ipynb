## Graphs(Plots)

```python
import matplotlib.pyplot as plt
import numpy as np
plt.title=("line plot")
plt.xlabel=("scale1")
plt.ylabel=("scale2")
xpoints=np.array([0,20])
ypoints=np.array([0,25])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
plt.plot(xpoints,ypoints,"--")
```

```python
xpoints=np.array([1,4,7,8,9])
ypoints=np.array([1,4,7,8,9])
plt.scatter(xpoints,ypoints,marker="*",color="b")
plt.show()
```

```python
xpoints=np.array([0,1,5,6,8])
ypoints=np.array([5,7,8,11,15])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints=np.array([0,1,5,6,8])
ypoints=np.array([5,7,8,11,15])
plt.plot(xpoints,ypoints,marker='o',mec='b',color='r',linewidth="2")
plt.show()
```
