## Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,1,3,4,5])
y=np.array([1,1,2,3,4,5])
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,s=100)
plt.show()
```

```python
plt.scatter(x,y,s=100,c="r",alpha=0.4)
plt.show()
```

## Bubble Plot

```python
x=np.random.rand(40)
print(x)
y=np.random.rand(40)
print(y)
z=np.random.rand(40)
print(y)
color=np.random.rand(40)
plt.scatter(x,y,z*1000,c=color*100,alpha=0.8)
plt.show()
```
